```python
idx = 0
cdk_plugin = get_cdk_plugin()
original_smiles = "CN(C)C\C=C\C(=O)Nc3cc1c(Nc(cc2Cl)ccc2F)ncnc1cc3OC4COCC4"
kwargs = {
    "is_dash" : True, 
    "is_wave_line" : False, 
    "dash_line" : False
}


image, result_dict = random_generate_rgroup_type_2(
                                                    idx = idx,
                                                    original_smiles = original_smiles,
                                                    cdk_plugin = cdk_plugin,
                                                    **kwargs
                                                    )
image.save("image/rgroup_type_2_dash.png")
image
```

```python
from PIL import Image, ImageDraw, ImageFont
draw = ImageDraw.Draw(image)
font_size = 30
offset_size = font_size//3
font = ImageFont.truetype("font/Arial.ttf", font_size)
for i in range(len(result_dict["0"]["graph"]["coords"])):
    x, y = result_dict["0"]["graph"]["coords"][i]
    label = result_dict["0"]["graph"]["symbols"][i]
    draw.text((x-offset_size, y-offset_size), label, fill="red", font=font)
image.save("image/rgroup_type_2_dash_plot.png")
image
```
